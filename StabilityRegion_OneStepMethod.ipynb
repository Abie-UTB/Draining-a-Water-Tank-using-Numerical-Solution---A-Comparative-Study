```python
import numpy as np
import matplotlib.pyplot as plt

# Define the stability functions
def euler(z):
    return 1 + z

def rk2(z):
    return 1 + z + (z**2) / 2

def rk4(z):
    return 1 + z + (z**2) / 2 + (z**3) / 6 + (z**4) / 24

# Create a grid in the complex plane
x = np.linspace(-3, 3, 500)
y = np.linspace(-3, 3, 500)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Calculate the magnitude of the stability functions
euler_mag = np.abs(euler(Z))
rk2_mag = np.abs(rk2(Z))
rk4_mag = np.abs(rk4(Z))

# Plot the stability regions
plt.figure(figsize=(15, 5))

# Euler's method
plt.subplot(1, 3, 1)
plt.contourf(X, Y, euler_mag, levels=[0, 1], colors=['blue'], alpha=0.5)
plt.title("Euler's Method Stability Region")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)

# Second-order Runge-Kutta method
plt.subplot(1, 3, 2)
plt.contourf(X, Y, rk2_mag, levels=[0, 1], colors=['green'], alpha=0.5)
plt.title("Second-Order Runge-Kutta Stability Region")
plt.xlabel('Re(z)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)

# Fourth-order Runge-Kutta method
plt.subplot(1, 3, 3)
plt.contourf(X, Y, rk4_mag, levels=[0, 1], colors=['red'], alpha=0.5)
plt.title("Fourth-Order Runge-Kutta Stability Region")
plt.xlabel('Re(z)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the stability functions
def explicit_euler(z):
    return 1 + z

def implicit_euler(z):
    return 1 / (1 - z)

# Create a grid in the complex plane
x = np.linspace(-3, 3, 500)
y = np.linspace(-3, 3, 500)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Calculate the magnitude of the stability functions
explicit_euler_mag = np.abs(explicit_euler(Z))
implicit_euler_mag = np.abs(implicit_euler(Z))

# Plot the stability regions
plt.figure(figsize=(10, 5))

# Explicit Euler
plt.subplot(1, 2, 1)
plt.contourf(X, Y, explicit_euler_mag, levels=[0, 1], colors=['blue'], alpha=0.5)
plt.title("Explicit Euler Stability Region")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)

# Implicit Euler
plt.subplot(1, 2, 2)
plt.contourf(X, Y, implicit_euler_mag, levels=[0, 1], colors=['green'], alpha=0.5)
plt.title("Implicit Euler Stability Region")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
